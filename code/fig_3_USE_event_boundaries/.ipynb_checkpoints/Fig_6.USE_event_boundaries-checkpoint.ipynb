### This script is being used to plan the partial mantel to run in R and then plot:
- FDR corrected by movie
- p<.05 across movies
- FDR corrected + 3 ROIs that are p <.05 in the social movies

```python
import numpy as np
import pandas as pd
import sys
import os
import matplotlib.pyplot as plt
import nibabel as nib
import mantel
from statsmodels.stats.multitest import multipletests
from nilearn.plotting import plot_glass_brain, plot_stat_map,  view_img, view_img_on_surf, plot_surf, plot_surf_stat_map, plot_surf_roi, plot_surf_contours, view_surf
from nilearn.surface import vol_to_surf
from nilearn import datasets
fsaverage = datasets.fetch_surf_fsaverage()
```

```python
def get_matchz_matrix(mov,node):
    schaeffer_dir = '/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/data/_schaeffer_rois/_comparisons_no_range_no_train/'

    matchz_node =  np.load(schaeffer_dir + "_matchz_individuals/"+ "%s/matchz_ind_44by44_node%s_%s_no_range_no_train.npy" %(mov,node,mov))

    return matchz_node


def get_motion(mov,method):
    ''' Method can be ISC or AnnaK'''
    motionISC_matrix = np.load(os.path.join(f'/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/linear_model_building/motion/motion{method}_matrix_{mov}.npy')) #using relative paths should prevent having to change path if not on discovery 
    
    return motionISC_matrix

def get_USE(mov):
    USE_matrix = np.load(os.path.join(f'/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/linear_model_building/USE/USE_matrix_{mov}.npy')) #using relative paths should prevent having to change path if not on discovery 
    
    return USE_matrix

def get_memory_AnnaK(mov):
    memoryAnnaK_matrix = np.load(os.path.join(f'/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/linear_model_building/memory/memoryAnnaK_matrix_{mov}.npy')) #using relative paths should prevent having to change path if not on discovery 
    
    return memoryAnnaK_matrix

def mean_center(lower_tri):
    ''' Make sure you are feeding in the lower triangle of the matrix'''
    return [val-np.nanmean(lower_tri) for val in lower_tri]

def drop_nans_for_mantel(x):
    where_nan = np.argwhere((np.isnan(x)))

    if len(where_nan) > 0:
        print('nan',node)
        for c in range(len(where_nan)):
            x[where_nan[c]] = np.nanmin(x)
    return x
    
```

```python
''' Step 0 - drop the subject from the matchz and motion matrices that does not have USE (no transcript - number 107)'''

''' Removing the subject that has no transcript'''
# usually would just see where there is an empty transcript, but for the purpose of this script, we know it is index 4/subject 107
nums = list(range(103,118))+list(range(119,121)) + list(range(122,131))+ list(range(132,137))+ list(range(138,139)) + list(range(140,141))+list(range(142,143)) + list(range(145,152))+ list(range(153,154)) + list(range(155,156))                                                                         
idx = nums.index(107)


def delete_sub4(matrix):
    if matrix.shape[0]==43:
        
        matrix = np.delete(matrix, (idx), axis=0)

        matrix = np.delete(matrix, (idx), axis=1)
    
    return matrix

def drop_nans_for_mantel(x):
    where_nan = np.argwhere((np.isnan(x)))

    if len(where_nan) > 0:
        print('nan',node)
        for c in range(len(where_nan)):
            x[where_nan[c]] = np.nanmin(x)
    return x
    
```

```python
mov = 'lemonade'
to_predict = 'USE'
predictors = ['matchz']
fixed_of_interest = 'matchz'

motionISC_matrix = get_motion(mov,'ISC')
motionISC_matrix = delete_sub4(motionISC_matrix)

motionAnnaK_matrix = get_motion(mov,'AnnaK')
motionAnnaK_matrix = delete_sub4(motionAnnaK_matrix)

memoryAnnaK_matrix = get_memory_AnnaK(mov)
memoryAnnaK_matrix = delete_sub4(memoryAnnaK_matrix)


USE_matrix = get_USE(mov)
low_t_inds = np.tril_indices(USE_matrix.shape[0], k=-1)


d = {'USE': list(USE_matrix[low_t_inds]), 
     'MotionISC': list(mean_center(motionISC_matrix[low_t_inds])), 
    'MotionAnnaK': list(mean_center(motionAnnaK_matrix[low_t_inds])), 
    'MemoryAnnaK': list(mean_center(memoryAnnaK_matrix[low_t_inds]))}
df = pd.DataFrame(d)
df.head()
dict_of_all_values = {}
for node in range(1,101):
    dict_of_all_values[node] = {}
    
for node in range(1,101):
    matchz = get_matchz_matrix(mov,node)
    matchz = drop_nans_for_mantel(matchz)
    matchz = delete_sub4(matchz)

    ''' True matrix'''
    df[f'matchz_{node}'] = list(mean_center(matchz[low_t_inds]))
    
df.to_csv(save_dir + '_csv_for_setup/' + f'{mov}_nodes_{to_predict}_{predictors[0]}.csv')
    
    
```

```python
def linear_model_USE(mov):

    save_dir = '/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/data/_schaeffer_rois/_comparisons_no_range_no_train/_linear_models/'
        #a = np.load(save_dir + f'USE+matchzMotionISCMotionAnnaKMemoryAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'USE+matchzMotionISCMotionAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'matchz+USEMotionISCMotionAnnaKMemoryAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'USE+matchz_{mov}_output_vals.npy',allow_pickle=True).item()
    #r_p_df = pd.read_csv(save_dir + '_csv_for_setup/' + f'{mov}_partial_mantel_r_p.csv')
    r_p_df = pd.read_csv(save_dir + '_csv_for_setup/' + f'{mov}_mantel_r_p_R_generated.csv')
    
        
    node_OG = r_p_df['rval']
    pval_list = r_p_df['pval']


    plotting_a = []
    listOf_reg = []
    val = []
    ''' Account for p-values calcuated with different order'''
    fdr_corrected_p = multipletests(pval_list,method='fdr_bh')[1]
#     print(fdr_corrected_p)

    plt.hist(pval_list)
    
    
    for i in range(100):
        if fdr_corrected_p[i] < .05:
            plotting_a.append(node_OG[i])
            val.append(node_OG[i])
            listOf_reg.append(i)
        else:
            plotting_a.append(np.nan)
    print(listOf_reg)
    #df = schaeffer_region_info(listOf_reg)
    #df['val'] = val
    return plotting_a

def get_node_r_val(rois,mov):
    ''' This function is intended to get the value for each node on its own '''
    save_dir = '/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/data/_schaeffer_rois/_comparisons_no_range_no_train/_linear_models/'
    r_p_df = pd.read_csv(save_dir + '_csv_for_setup/' + f'{mov}_mantel_r_p_R_generated.csv')
    node_OG = r_p_df['rval']
    
    vals = []
    for val in rois:
        vals.append(node_OG[val])
    return vals



def linear_model_USE_notFDR(mov):

    save_dir = '/dartfs/rc/lab/F/FinnLab/clara/K99_EventSeg/data/_schaeffer_rois/_comparisons_no_range_no_train/_linear_models/'
        #a = np.load(save_dir + f'USE+matchzMotionISCMotionAnnaKMemoryAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'USE+matchzMotionISCMotionAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'matchz+USEMotionISCMotionAnnaKMemoryAnnaK_{mov}_output_vals.npy',allow_pickle=True).item()
        #a = np.load(save_dir + f'USE+matchz_{mov}_output_vals.npy',allow_pickle=True).item()
    #r_p_df = pd.read_csv(save_dir + '_csv_for_setup/' + f'{mov}_partial_mantel_r_p.csv')
    r_p_df = pd.read_csv(save_dir + '_csv_for_setup/' + f'{mov}_mantel_r_p_R_generated.csv')
    
        
    node_OG = r_p_df['rval']
    pval_list = r_p_df['pval']


    plotting_a = []
    listOf_reg = []
    val = []
    ''' Account for p-values calcuated with different order'''
    
    
    for i in range(100):
        if pval_list[i] < .05:
            plotting_a.append(node_OG[i])
            val.append(node_OG[i])
            listOf_reg.append(i)
        else:
            plotting_a.append(np.nan)
    #df = schaeffer_region_info(listOf_reg)
    #df['val'] = val
    return plotting_a,listOf_reg


def color_rois(values,data_dir):
    """
    This function assumes you are passing a vector "values" with the same length as the number of nodes in the atlas.
    """
    
    atlas_fname = (data_dir + '_masks/'+ 'Schaefer2018_100Parcels_7Networks_order_FSLMNI152_1mm.nii.gz')
    schaeffer = nib.load(atlas_fname)
    schaeffer100_data = schaeffer.get_fdata()
    img = np.zeros(schaeffer100_data.shape)
    for roi in range(len(values)):
        itemindex = np.where(schaeffer100_data==roi+1) # find voxels in this node (subtract 1 to account for zero-indexing)
        img[itemindex] = values[roi] # color them by the desired value 
    affine = schaeffer.affine
    img_nii = nib.Nifti1Image(img, affine)
    
    return img_nii
def plot_isrsa(isrsa_mantel_r2_brain,colorbar,vmax,vmin):
    cmap='RdBu_r'
#     vmin = -.17
#     vmax = .17
    fig,ax = plt.subplots(nrows=1, ncols= 2, figsize=(8,8),subplot_kw={'projection': '3d'})
    fig.tight_layout(h_pad=None,w_pad=None)
    texture = vol_to_surf(isrsa_mantel_r2_brain, fsaverage.pial_left,interpolation='nearest',radius =1, n_samples=1)
    surf_plot1=plot_surf_roi(fsaverage.infl_left, texture, hemi='left',cmap = cmap, colorbar=False, vmin=vmin,  vmax = vmax,
                                 bg_map=fsaverage.sulc_left,axes=ax[0],darkness=.4)
    texture = vol_to_surf(isrsa_mantel_r2_brain, fsaverage.pial_right,interpolation='nearest',radius =1, n_samples=1)
    surf_plot3=plot_surf_roi(fsaverage.infl_right, texture, hemi='right',cmap = cmap, colorbar=False,vmin=vmin,  vmax = vmax,
                                 bg_map=fsaverage.sulc_right,axes=ax[1],darkness=.4)


    fig,ax = plt.subplots(nrows=1, ncols= 2, figsize=(8,8),subplot_kw={'projection': '3d'})
    fig.tight_layout(h_pad=None,w_pad=None)
    texture = vol_to_surf(isrsa_mantel_r2_brain, fsaverage.pial_left,interpolation='nearest',radius =1, n_samples=1)
    surf_plot2=plot_surf_roi(fsaverage.infl_left, texture, hemi='left',cmap = cmap, colorbar=False,vmin=vmin,  vmax = vmax,
                                bg_map=fsaverage.sulc_left, view = 'medial',axes=ax[1],darkness=.4)
    texture = vol_to_surf(isrsa_mantel_r2_brain, fsaverage.pial_right,interpolation='nearest',radius =1, n_samples=1)
    surf_plot4=plot_surf_roi(fsaverage.infl_right, texture, hemi='right',cmap = cmap, colorbar=False,vmin=vmin,  vmax = vmax,
                                bg_map=fsaverage.sulc_right, view ='medial',axes=ax[0],darkness=.4)
    
    
    if colorbar==True:
        plot_stat_map(isrsa_mantel_r2_brain,vmax=vmax,symmetric_cbar=True,cmap = cmap)
    

```
